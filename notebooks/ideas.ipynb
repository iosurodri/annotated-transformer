```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib widget
```

```python
import os
PATH_IMAGES = os.path.join('..', 'images')
```

```python
def penality_similarity(x, y, alpha=0.5, beta=0.5, theta=0.5):
    ref_xy = 1 - np.abs(x - y)
    dist_x_theta = np.abs(x - theta)
    dist_y_theta = np.abs(y - theta)
    return alpha * ref_xy + beta * (dist_x_theta + dist_y_theta)

def a1_to_01(x, a, theta=1):
    return ((x - a) / (1 - a)) ** theta
```

```python
alpha = 0.5
beta = 1 - alpha
x = np.arange(0, 1.01, 0.05)
y = np.arange(0, 1.01, 0.05)
X, Y = np.meshgrid(x, y)
Z = penality_similarity(X, Y, alpha=alpha, beta=beta)
Z = a1_to_01(Z, min(alpha, beta))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
simm_surface = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='coolwarm', edgecolor='none')
lower_lim = ax.plot_surface(X, Y, np.ones(X.shape) * np.min(Z), rstride=1, cstride=1, cmap='coolwarm', edgecolor='none', alpha=0.3)
ax.set_xlabel('x')
ax.set_xlim(0, 1)
ax.set_ylabel('y')
ax.set_ylim(0, 1)
ax.set_zlabel('z')
ax.set_zlim(0, 1)
plt.title(r'Penalty simm: $\alpha$={}, $\beta$={}'.format(alpha, 1-alpha))
# plt.savefig(os.path.join(PATH_IMAGES, 'penalized_sim_drastic_alpha_{}_beta_{}.png'.format(str(alpha), str(1-alpha))))
fig.colorbar(simm_surface, shrink=0.5, aspect=5)
```
